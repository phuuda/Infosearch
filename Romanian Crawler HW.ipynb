```python
# crawler собрать все ссылки со страниц румынской википедии
import requests, re
from bs4 import BeautifulSoup
from IPython.display import HTML, display

all_links = []
links_checked = []

def romanian_crawl(query):
    
    req = requests.get(query) # https://mo.wikipedia.org/
    soup = BeautifulSoup(req.text, 'lxml')

    for m in soup.findAll('a'):
    
        if m.has_attr('href'):
            link = m['href']
            if link.startswith('/wiki/%') or link.startswith('/wiki/Categorie:'):
                
                match = re.search('redlink', link)
                if not match:

                    link = 'https://mo.wikipedia.org' + link
                    if link not in all_links and link not in links_checked:
                        all_links.append(link)

def recrawl(query):   
    while all_links:
        
        for el in all_links:
            if el not in links_checked:
                romanian_crawl(el)
                links_checked.append(el)
                all_links.remove(el)
                
            else:
                all_links.remove(el)
                
            #print(len(all_links), len(links_checked))
            
romanian_crawl('https://mo.wikipedia.org/')    
recrawl('https://mo.wikipedia.org/') # this one takes about 5 minutes
```
